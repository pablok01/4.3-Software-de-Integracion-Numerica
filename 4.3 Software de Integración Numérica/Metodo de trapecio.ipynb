```python
import math
import numpy as np
def evaluacion(x):
  y=np.exp(2*x)*math.sin(3*x)
  return y
b=2
a=0
n=8
h=(b-a)/n
count=0
i=1
suma=0
print('{:^10}{:^10}{:^10}'.format('i','xi','evaluacion'))
print('{:^10}{:^10}{:^10}'.format('--------------','--------------','--------------'))
print('{:^10}{:^10}{:^10}'.format('0',a,evaluacion(a)))
while i<n:
  xi=a+i*h
  count+=1
  suma+=evaluacion(xi)
  i+=1
  print('{:^10}{:^10.1f}{:^10.4f}'.format(count,xi,evaluacion(xi)))
resultado=h/2*(evaluacion(a)+evaluacion(b))+h*suma
print('{:^10}{:^10}{:^10}'.format(n,b,evaluacion(b)))
print()
print("la solucion es: ",resultado)
```
